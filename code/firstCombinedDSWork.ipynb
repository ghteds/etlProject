```python
#were still doing leg work to one file at a 
#time building to the creshendo and the loop that will loop through files
thePath='../datasets/Combined3StatesExcel/'
```

```python
#using glob to help us loop through files and grab all files
import glob
```

```python
#this one will loop through all files...grab the needed data and store in summary

theFiles=glob.glob(thePath + '*.csv')
#new dictionary to store auto dfs; this is really nice cuz 
#the dfs can be accessed with simple syntax even though stored in dictionary(eg df[key]or in this case theDFs[Year])
#these are temporary dfs though with the final df being the summary df
theDFs={}
states=['Georgia','Ohio','Oregon']
for afile in theFiles:
    #     print(afile)
    #get the year fromt the file names
    #grab the year from the file name to use in the name/key in dictionary
    theYear=os.path.splitext(afile)[0][-4:]
    
    #use the year in the df name/key
    #this is pretty cool to auto create and store and easily reference new dfs store dict
    theDFs[theYear]=pd.read_csv(afile, header=1)
    #TRANSPOSE or flip the rows and cols
    theDFs[theYear] = theDFs[theYear].T
    #set the col names to the states from row 3
    theDFs[theYear].columns = theDFs[theYear].iloc[2]
    #print(theDFs[theYear].head())
    #---------------------------------------------------------------
  
    
    for state in states:
        #state year used to column name... need to build to reference column
        stateYear = state + theYear
        #print(stateYear)
        #name of state
        #print(f'{state}',state,theYear)
        #TOTAL
        theTotal=theDFs[theYear].loc['Estimate; Total:'][f'{state}']
        stateSummaryDf.loc['Total',stateYear]=theTotal
        #print(theTotal)
       #UNINSURED
        theUninsured=theDFs[theYear][(theDFs[theYear].index.str.contains('Estimate')) & \
         (theDFs[theYear].index.str.contains('No health insurance coverage'))][f'{state}'].sum()
        stateSummaryDf.loc['Uninsured',stateYear]=theUninsured
        #print(theUninsured)
        #RATE
        theRate=theUninsured/theTotal
        stateSummaryDf.loc['Rate',stateYear]=theRate
        #print(theRate)

stateSummaryDf.head()
    
```
